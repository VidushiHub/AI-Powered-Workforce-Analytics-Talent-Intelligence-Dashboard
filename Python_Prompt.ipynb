## Student Management System with Pandas

```python
# Import the pandas library
import pandas as pd
```

```python
# 1. Create a sample DataFrame for student details
# In a real scenario, you might read this from a CSV, Excel, or database.
data = {
    'Student ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [85, 92, 78, 95, 60],
    'Science': [90, 88, 85, 91, 75],
    'English': [75, 80, 92, 88, 70]
}

df = pd.DataFrame(data)
display(df)
```

```python
# 2. Calculate the average marks for each student
# We will calculate the average across 'Math', 'Science', and 'English' columns.
df['Average Marks'] = df[['Math', 'Science', 'English']].mean(axis=1)
display(df)
```

```python
# 3. Find the highest scorer based on 'Average Marks'
highest_scorer = df.loc[df['Average Marks'].idxmax()]
print("\nHighest Scorer:")
display(highest_scorer)
```

```python
# 4. Display students scoring above 80 in 'Average Marks'
students_above_80 = df[df['Average Marks'] > 80]
print("\nStudents scoring above 80 in Average Marks:")
display(students_above_80)
```

```python
# 5. Save the DataFrame to a CSV file
# The index=False argument prevents pandas from writing the DataFrame index as a column in the CSV.
output_filename = 'student_data.csv'
df.to_csv(output_filename, index=False)
print(f"\nStudent data saved to {output_filename}")

# You can verify the content by reading it back
# verification_df = pd.read_csv(output_filename)
# display(verification_df.head())
```
