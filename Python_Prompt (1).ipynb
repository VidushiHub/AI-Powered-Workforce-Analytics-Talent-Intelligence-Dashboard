## Student Management System with Pandas

```python
# Import the pandas library
import pandas as pd
```

```python
# 1. Create a sample DataFrame for student details
# In a real scenario, you might read this from a CSV, Excel, or database.
data = {
    'Student ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math': [85, 92, 78, 95, 60],
    'Science': [90, 88, 85, 91, 75],
    'English': [75, 80, 92, 88, 70]
}

df = pd.DataFrame(data)
display(df)
```

```python
# 2. Calculate the average marks for each student
# We will calculate the average across 'Math', 'Science', and 'English' columns.
df['Average Marks'] = df[['Math', 'Science', 'English']].mean(axis=1)
display(df)
```

```python
# 3. Find the highest scorer based on 'Average Marks'
highest_scorer = df.loc[df['Average Marks'].idxmax()]
print("\nHighest Scorer:")
display(highest_scorer)
```

```python
# 4. Display students scoring above 80 in 'Average Marks'
students_above_80 = df[df['Average Marks'] > 80]
print("\nStudents scoring above 80 in Average Marks:")
display(students_above_80)
```

```python
# 5. Save the DataFrame to a CSV file
# The index=False argument prevents pandas from writing the DataFrame index as a column in the CSV.
output_filename = 'student_data.csv'
df.to_csv(output_filename, index=False)
print(f"\nStudent data saved to {output_filename}")

# You can verify the content by reading it back
# verification_df = pd.read_csv(output_filename)
# display(verification_df.head())
```

### Lists
Lists are ordered, mutable collections of items. They are one of the most versatile data types in Python.

```python
# 1. Creating a list
my_list = [1, 2, 3, 'apple', 'banana', True]
print(f"Original list: {my_list}")

# 2. Accessing elements (indexing)
print(f"First element: {my_list[0]}")
print(f"Last element: {my_list[-1]}")

# 3. Slicing a list
print(f"Slice from index 1 to 3 (exclusive): {my_list[1:4]}")

# 4. Modifying elements (lists are mutable)
my_list[0] = 100
print(f"List after modifying first element: {my_list}")

# 5. Adding elements
my_list.append('grape') # Add to the end
print(f"List after appending 'grape': {my_list}")
my_list.insert(1, 'orange') # Insert at specific index
print(f"List after inserting 'orange' at index 1: {my_list}")

# 6. Removing elements
my_list.remove('banana') # Remove by value
print(f"List after removing 'banana': {my_list}")
popped_item = my_list.pop() # Remove and return last item
print(f"List after popping last item ({popped_item}): {my_list}")

# 7. Iterating through a list
print("\nIterating through the list:")
for item in my_list:
    print(item)
```

### Tuples
Tuples are ordered, immutable collections of items. Once created, their elements cannot be changed.

```python
# 1. Creating a tuple
my_tuple = (1, 2, 3, 'cat', 'dog', False)
print(f"Original tuple: {my_tuple}")

# 2. Accessing elements (indexing)
print(f"First element: {my_tuple[0]}")
print(f"Last element: {my_tuple[-1]}")

# 3. Slicing a tuple
print(f"Slice from index 0 to 2 (exclusive): {my_tuple[0:3]}")

# 4. Tuples are immutable (this will raise an error if uncommented)
# my_tuple[0] = 100 # Uncommenting this will cause a TypeError
# print(f"Tuple after attempt to modify: {my_tuple}")

# 5. Concatenating tuples
another_tuple = ('fish', 'bird')
combined_tuple = my_tuple + another_tuple
print(f"Combined tuple: {combined_tuple}")

# 6. Iterating through a tuple
print("\nIterating through the tuple:")
for item in my_tuple:
    print(item)
```

### Loops
Loops are used to execute a block of code repeatedly. Python supports `for` loops and `while` loops.

```python
# 1. For loop with a list
fruits = ['apple', 'banana', 'cherry']
print("\nFor loop through a list:")
for fruit in fruits:
    print(fruit)

# 2. For loop with range()
print("\nFor loop using range(5):")
for i in range(5): # Iterates from 0 to 4
    print(i)

# 3. While loop
print("\nWhile loop example:")
count = 0
while count < 3:
    print(f"Count: {count}")
    count += 1

# 4. Loop with 'break'
print("\nFor loop with 'break':")
for i in range(10):
    if i == 5:
        print("Breaking loop at 5")
        break
    print(i)

# 5. Loop with 'continue'
print("\nFor loop with 'continue':")
for i in range(5):
    if i == 2:
        print("Skipping 2")
        continue # Skip the rest of the current iteration
    print(i)
```

### Object-Oriented Programming (OOP) Concepts
OOP is a programming paradigm based on the concept of "objects", which can contain data and code.

```python
# 1. Defining a class
class Dog:
    # Class attribute
    species = "Canis familiaris"

    # Constructor (__init__) method
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    # Instance method
    def bark(self):
        return f"{self.name} says Woof!"

    # Another instance method
    def description(self):
        return f"{self.name} is {self.age} years old."

# 2. Creating objects (instances of the class)
my_dog = Dog("Buddy", 3)
your_dog = Dog("Lucy", 5)

# 3. Accessing attributes
print(f"My dog's name: {my_dog.name}")
print(f"Your dog's age: {your_dog.age}")
print(f"Species of dogs: {Dog.species}")

# 4. Calling methods
print(f"My dog's sound: {my_dog.bark()}")
print(f"Your dog's description: {your_dog.description()}")

# 5. Modifying instance attributes
my_dog.age = 4
print(f"My dog's new age: {my_dog.age}")
```
